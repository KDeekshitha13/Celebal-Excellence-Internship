**Step-3:Handling Missing values**

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['videos'])
```

```python
numerical_cols=df.select_dtypes(include=['number']).columns
numerical_cols
df[numerical_cols]=df[numerical_cols].fillna(df[numerical_cols].median())
```

```python
text_cols=df.select_dtypes(include=['object']).columns
text_cols
df[text_cols]=df[text_cols].fillna("Unknown")
```

```python
df.isnull().sum()
```

The videos column was removed from the dataset because a large portion of its values (approximately 78.1%) were missing, making the column unreliable for analysis. To handle the remaining missing data, the median was used to fill missing values in numeric columns, as it is less affected by extreme values. For categorical and text-based columns, missing entries were replaced with "Unknown", ensuring that valuable records were retained while maintaining the overall integrity of the dataset.

**Step 4:Basic Operations**

```python
selected_data=df[['title','rating','final_price','category']]
selected_data
```

```python
high_rating=df[df['rating']>4]
high_rating.head()
```

```python
df['final_price']
```

```python
df['final_price']=df['final_price'].str.replace('₹','',regex=False)
df['final_price']=df['final_price'].str.replace(',','',regex=False)
df['final_price']=df['final_price'].str.replace('"','',regex=False)
df['final_price']=df['final_price'].astype(float)
```

```python
print(df['final_price'].dtype)
```

```python
expensive_products=df[df['final_price']>3000]
expensive_products.head()
```

Basic data manipulation tasks were carried out using Pandas to better understand and analyze the dataset. Key columns such as product title, rating, final price, and category were selected to focus on the most relevant product details. The data was then filtered to identify products with ratings above 4 as well as products with higher final prices, allowing for a more targeted analysis. These operations helped uncover useful insights from the dataset and demonstrated how specific subsets of data can be extracted efficiently for further exploration and decision-making.
